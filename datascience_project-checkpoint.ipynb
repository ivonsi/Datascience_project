## Periodi di alta/bassa richiesta

```python
import pandas as pd

FILE_PATH = "eCheckin_Guests_2024.xlsx"

def count_nights(df: pd.DataFrame, date_start: str, date_end: str, month: str):
    df["check_in"] = pd.to_datetime(df["touristStayFrom"], format="%d.%m.%Y %H:%M:%S", errors="coerce")
    df["check_out"] = pd.to_datetime(df["touristForseenStayUntil"], format="%d.%m.%Y %H:%M:%S", errors="coerce")

    start_date = pd.to_datetime(date_start)
    end_date = pd.to_datetime(date_end)

    df_filtered = df[
        (df["check_in"] < end_date) & (df["check_out"] > start_date)
    ].copy()

    notti_occupate = set()

    for _, row in df_filtered.iterrows():
        if pd.notnull(row["check_in"]) and pd.notnull(row["check_out"]):
            inizio = max(row["check_in"].normalize(), start_date)
            fine = min(row["check_out"].normalize(), end_date)

            for giorno in pd.date_range(inizio, fine - pd.Timedelta(days=1)):
                key = (row["apartmentNumber"], giorno)
                notti_occupate.add(key)
                
    num_notti = len(notti_occupate)  
    
    print(f"Totali notti occupati a {month} 2024: {len(notti_occupate)}")

    return num_notti

df = pd.read_excel(
    FILE_PATH,
    sheet_name="Popis gostiju",
    skiprows=14
)

notti_giugno = count_nights(df, "2024-06-01", "2024-07-01", "giugno")
notti_luglio = count_nights(df, "2024-07-01", "2024-08-01", "luglio")
notti_agosto = count_nights(df, "2024-08-01", "2024-09-01", "agosto")
notti_settembre = count_nights(df, "2024-09-01", "2024-10-01", "settembre")

dates = set()
for _, row in df.iterrows():
    dates.add((row["apartmentNumber"], row["touristStayFrom"], row["numberOfNights"]))

total = 0
for *_, n in dates:
    total += n

print("TOTALE:", total)  # 262

print("\nPercentuali:")
print(f"Giugno: {notti_giugno/total*100:.1f}%")
print(f"Luglio: {notti_luglio/total*100:.1f}%")
print(f"Agosto: {notti_agosto/total*100:.1f}%")
print(f"Settembre: {notti_settembre/total*100:.1f}%")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel(
    FILE_PATH,
    sheet_name="Popis gostiju",
    skiprows=14
)

df["check_in"] = pd.to_datetime(df["touristStayFrom"], format="%d.%m.%Y %H:%M:%S", errors="coerce")
df["check_out"] = pd.to_datetime(df["touristForseenStayUntil"], format="%d.%m.%Y %H:%M:%S", errors="coerce")

def notti_occupate_mese(df, anno, mese):
    start_date = pd.Timestamp(f"{anno}-{mese:02d}-01")
    if mese == 12:
        end_date = pd.Timestamp(f"{anno+1}-01-01")
    else:
        end_date = pd.Timestamp(f"{anno}-{mese+1:02d}-01")

    df_filtered = df[
        (df["check_in"] < end_date) & (df["check_out"] > start_date)
    ].copy()

    notti = set()

    for _, row in df_filtered.iterrows():
        if pd.notnull(row["check_in"]) and pd.notnull(row["check_out"]):
            inizio = max(row["check_in"].normalize(), start_date)
            fine = min(row["check_out"].normalize(), end_date)

            for giorno in pd.date_range(inizio, fine - pd.Timedelta(days=1)):
                key = (row["apartmentNumber"], giorno)
                notti.add(key)

    return len(notti)

mesi_italiano = ['Gennaio', 'Febbraio', 'Marzo', 'Aprile', 'Maggio', 'Giugno',
                 'Luglio', 'Agosto', 'Settembre', 'Ottobre', 'Novembre', 'Dicembre']

notti_per_mese = [notti_occupate_mese(df, 2024, m) for m in range(1, 13)]

plt.figure(figsize=(12, 6))
bars = plt.bar(mesi_italiano, notti_per_mese, color='lightblue')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom')

plt.yticks(range(0, 121, 20))
plt.ylim(0, 120)

plt.title('Notti occupate per mese (2024)', fontsize=14)
plt.xlabel('Mese', fontsize=12)
plt.ylabel('Notti occupate', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

## Confronto tra i 4 appartamenti

```python
import pandas as pd
from collections import defaultdict

df = pd.read_excel(
    FILE_PATH,
    sheet_name="Popis gostiju",
    skiprows=14
)

df["check_in"] = pd.to_datetime(df["touristStayFrom"], format="%d.%m.%Y %H:%M:%S", errors="coerce")
df["check_out"] = pd.to_datetime(df["touristForseenStayUntil"], format="%d.%m.%Y %H:%M:%S", errors="coerce")

df = df.dropna(subset=["check_in", "check_out", "apartmentNumber"])

tutti_appartamenti = sorted(df["apartmentNumber"].dropna().unique())

def calcola_occupazione_per_mese(df, anno, mese, nome_mese, appartamenti):
    start_date = pd.Timestamp(f"{anno}-{mese:02d}-01")
    if mese == 12:
        end_date = pd.Timestamp(f"{anno + 1}-01-01")
    else:
        end_date = pd.Timestamp(f"{anno}-{mese + 1:02d}-01")

    df_mese = df[(df["check_in"] < end_date) & (df["check_out"] > start_date)].copy()

    occupazione = defaultdict(set)

    for _, row in df_mese.iterrows():
        inizio = max(row["check_in"].normalize(), start_date)
        fine = min(row["check_out"].normalize(), end_date)

        for giorno in pd.date_range(inizio, fine - pd.Timedelta(days=1)):
            occupazione[row["apartmentNumber"]].add(giorno)

    print(f"\nNOTTI OCCUPATE A {nome_mese.upper()} 2024 PER APPARTAMENTO:")
    max_notti = 0
    risultati = {}

    for apt in appartamenti:
        notti = len(occupazione[apt]) if apt in occupazione else 0
        risultati[apt] = notti
        print(f"  Appartamento {apt}: {notti} notti")
        max_notti = max(max_notti, notti)

    top_appartamenti = [apt for apt, n in risultati.items() if n == max_notti and max_notti > 0]

    if top_appartamenti:
        top_str = ", ".join(str(a) for a in top_appartamenti)
        print(f"APPARTAMENTO/I PIÙ OCCUPATO/I A {nome_mese.upper()}: {top_str} ({max_notti} notti)")
    else:
        print(f"Nessun appartamento occupato a {nome_mese}")

mesi_da_calcolare = [
    (6, "Giugno"),
    (7, "Luglio"),
    (8, "Agosto"),
    (9, "Settembre")
]

for mese, nome in mesi_da_calcolare:
    calcola_occupazione_per_mese(df, 2024, mese, nome, tutti_appartamenti)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
import numpy as np

df = pd.read_excel(
    FILE_PATH,
    sheet_name="Popis gostiju",
    skiprows=14
)

df["check_in"] = pd.to_datetime(df["touristStayFrom"], format="%d.%m.%Y %H:%M:%S", errors="coerce")
df["check_out"] = pd.to_datetime(df["touristForseenStayUntil"], format="%d.%m.%Y %H:%M:%S", errors="coerce")

df = df.dropna(subset=["check_in", "check_out", "apartmentNumber"])

tutti_appartamenti = sorted(df["apartmentNumber"].dropna().unique())

mesi_da_calcolare = [
    (6, "Giugno"),
    (7, "Luglio"),
    (8, "Agosto"),
    (9, "Settembre")
]

dati_plot = {apt: [] for apt in tutti_appartamenti}  

for mese_num, nome_mese in mesi_da_calcolare:
    start_date = pd.Timestamp(f"2024-{mese_num:02d}-01")
    end_date = pd.Timestamp(f"2024-{mese_num + 1:02d}-01") if mese_num < 12 else pd.Timestamp("2025-01-01")

    df_mese = df[(df["check_in"] < end_date) & (df["check_out"] > start_date)].copy()

    occupazione = defaultdict(set)

    for _, row in df_mese.iterrows():
        inizio = max(row["check_in"].normalize(), start_date)
        fine = min(row["check_out"].normalize(), end_date)
        for giorno in pd.date_range(inizio, fine - pd.Timedelta(days=1)):
            occupazione[row["apartmentNumber"]].add(giorno)

    for apt in tutti_appartamenti:
        notti = len(occupazione[apt]) if apt in occupazione else 0
        dati_plot[apt].append(notti)

mesi_nomi = [nome for _, nome in mesi_da_calcolare]
n_mesi = len(mesi_nomi)
n_app = len(tutti_appartamenti)

fig, ax = plt.subplots(figsize=(12, 6))
bar_width = 0.8 / n_app  
x_base = np.arange(n_mesi)
colori = plt.cm.plasma(np.linspace(0, 1, n_app))

for i, apt in enumerate(tutti_appartamenti):
    notti = dati_plot[apt]
    x_pos = x_base + i * bar_width
    bars = ax.bar(x_pos, notti, width=bar_width, color=colori[i], label=f"Apt {apt}")

    for bar in bars:
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width() / 2, h + 0.5, str(h), ha='center', va='bottom', fontsize=8)

ax.set_xticks(x_base + (bar_width * n_app / 2) - bar_width / 2)
ax.set_xticklabels(mesi_nomi)
ax.set_ylabel("Notti occupate")
ax.set_title("Notti occupate per mese e numero di appartamento (2024)")
ax.set_ylim(0, max(max(dati_plot[apt]) for apt in tutti_appartamenti) + 5)
ax.legend(title="Appartamenti", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_excel(FILE_PATH, sheet_name="Statistika po državama", skiprows=8)

traduzioni = {
    'Njemačka': 'Germania',
    'Poljska': 'Polonia',
    'Hrvatska': 'Croazia',
    'Kazahstan': 'Kazakistan',
    'Italija': 'Italia',
    'Slovenija': 'Slovenia',
    'Nizozemska': 'Paesi Bassi',
    'Austrija': 'Austria',
    'Mađarska': 'Ungheria',
    'Ukrajina': 'Ucraina',
    'Bosna i Hercegovina': 'Bosnia ed Erzegovina',
    'Slovačka': 'Slovacchia',
    'Bjelorusija': 'Bielorussia',
    'Grčka': 'Grecia',
    'Češka': 'Repubblica Ceca',
    'Srbija': 'Serbia',
    'Španjolska': 'Spagna'
}

df['Nazionalità'] = df['Citizenship'].map(traduzioni).fillna(df['Citizenship'])

def plot_visitors(df, column, ylabel, title):
    df_sorted = df.sort_values(by=column, ascending=False).copy()
    total = df_sorted[column].sum()
    df_sorted['Percentuale'] = df_sorted[column] / total * 100

    plt.figure(figsize=(12, 6))  

    bars = plt.bar(df_sorted['Nazionalità'], df_sorted[column], color='lightblue')
    max_height = df_sorted[column].max()
    plt.ylim(0, max_height * 1.15)  

    for bar, perc in zip(bars, df_sorted['Percentuale']):
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2,
            height + max_height * 0.02,  
            f"{perc:.1f}%",
            ha='center',
            va='bottom',
            fontsize=10
        )

    plt.xlabel('Nazionalità')
    plt.ylabel(ylabel)
    plt.title(title)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

plot_visitors(df, 'Count', 'Numero di visitatori', 'Distribuzione dei visitatori per nazionalità')

plot_visitors(df, 'numberOfNights', 'Numero di notti', 'Distribuzione delle notti per nazionalità')
```

## Confronto con l' anno 2023

```python
import pandas as pd
import matplotlib.pyplot as plt

mesi = {5: "Maggio", 6: "Giugno", 7: "Luglio", 8: "Agosto", 9: "Settembre"}

def calculate_room_nights(df, year_label):
    df["check_in"] = pd.to_datetime(df["touristStayFrom"], format="%d.%m.%Y %H:%M:%S", errors="coerce")
    df["check_out"] = pd.to_datetime(df["touristForseenStayUntil"], format="%d.%m.%Y %H:%M:%S", errors="coerce")
    df = df.dropna(subset=["check_in", "check_out", "apartmentNumber"])
    
    results = []
    for month in [5, 6, 7, 8, 9]:
        start_date = pd.Timestamp(f"{year_label}-{month:02d}-01")
        end_date = pd.Timestamp(f"{year_label}-{month + 1:02d}-01") if month < 12 else pd.Timestamp(f"{year_label+1}-01-01")
        
        df_month = df[(df["check_in"] < end_date) & (df["check_out"] > start_date)].copy()
        occupati = set()
        
        for _, row in df_month.iterrows():
            inizio = max(row["check_in"].normalize(), start_date)
            fine = min(row["check_out"].normalize(), end_date)
            
            for giorno in pd.date_range(inizio, fine - pd.Timedelta(days=1)):
                occupati.add((row["apartmentNumber"], giorno))
        
        results.append({
            "Anno": year_label,
            "Mese": mesi[month],
            "Notti Occupate": len(occupati)
        })
    
    return pd.DataFrame(results)

df_2023 = pd.read_excel("eCheckin_Guests_2023.xlsx", sheet_name="Popis gostiju", skiprows=14)
df_2024 = pd.read_excel("eCheckin_Guests_2024.xlsx", sheet_name="Popis gostiju", skiprows=14)

df_rn_2023 = calculate_room_nights(df_2023, 2023)
df_rn_2024 = calculate_room_nights(df_2024, 2024)

df_finale = pd.concat([df_rn_2023, df_rn_2024])
pivot_df = df_finale.pivot(index="Mese", columns="Anno", values="Notti Occupate")
pivot_df = pivot_df.loc[["Maggio", "Giugno", "Luglio", "Agosto", "Settembre"]]

plt.figure(figsize=(10, 6))
for year in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[year], marker='o', label=str(year))

plt.title("Confronto occupazione - 2023 vs 2024")
plt.xlabel("Mese")
plt.ylabel("Notti occupate")
plt.legend(title="Anno")
plt.grid(True)
plt.tight_layout()
plt.show()
```
